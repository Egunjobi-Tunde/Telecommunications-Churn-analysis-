# DATA CLEANING FOR MODEL PURPOSE

```python
data.isna().sum()
```

```python
# The following columns will not be needed for model building
Modata = data.drop(['Churn Category','Churn Reason', 'Customer ID', 'Zip Code', 'Latitude', 'Longitude', 'City', 'Number of Referrals'], axis = 1)
```

```python
Modata.head(4)
```

```python
Modata.isna().sum()
```

```python
Modata.dtypes
```

```python
Modata = Modata.fillna(0)
```

```python
Modata.isna().sum()
```

### All the columns with Object data type will be converted to integer for model construction purpose

```python
def col(Modata):
    for column in Modata:
        if Modata[column].dtype == 'object':
            print(f" {column} : {Modata[column].unique()}")
```

```python
col(Modata)
```

```python
Modata.replace('Joined', 'Stayed', inplace = True)
#Modata.fillna('No', inplace = True)
```

```python
col(Modata)
```

```python
yes_no_columns = ['Unlimited Data','Streaming Music', "Phone Service",'Online Security','Online Backup',
                 'Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Paperless Billing','Married','Internet Service',
                 'Multiple Lines']

for col in yes_no_columns:
    Modata[col].replace({'Yes': 1, 'No': 0}, inplace = True)
```

```python
for col in Modata:
    print(f'{col} : {Modata[col].unique()}')
```

```python
df = pd.get_dummies(columns = ['Gender', 'Customer Status','Payment Method','Contract','Internet Type','Offer'], data = Modata, drop_first = True)
df.head()
```

```python
df.columns.unique()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df['Customer Status_Stayed'].value_counts()
```

### The data was splitted into two, the first 5000 rows will be use to train the model while the remaining 2000+ rows will be use as a test data on the model to see how accurate the model is with a new data.

```python
train = df[:5000]
test = df[5000:]
```

```python
train.head(2)
```

```python
test.head(2)
```

```python
print(train['Customer Status_Stayed'].value_counts())
test['Customer Status_Stayed'].value_counts()
```

```python
X = train.drop('Customer Status_Stayed', axis = 1)
y = train['Customer Status_Stayed']

#test
testdata = test.drop('Customer Status_Stayed', axis = 1)
testoutput = test['Customer Status_Stayed']
```

```python
from imblearn.over_sampling import SMOTE
over = SMOTE()
X,y = over.fit_resample(X,y)
testdata, testoutput = over.fit_resample(testdata, testoutput)
```

```python
y.value_counts(), testoutput.value_counts() #the imbalance has been taking care
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 5)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## MODEL

```python
logmodel = LogisticRegression()
logmodel = logmodel.fit(X,y)
y_pred = logmodel.predict(X_test)
accuracy_score(y_test,y_pred)*100
```

```python
print(classification_report(y_test,y_pred))
```

# The Accuracy Score of the Logistic regression Model is 74%

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True, fmt = 'd')
plt.xlabel("Actual")
plt.ylabel("Prediction")
```

## Using the test data to check the accuracy of the Model

```python
y_pred_test = logmodel.predict(testdata)
accuracy_score(testoutput,y_pred_test)*100
```

```python
print(classification_report(testoutput,y_pred_test))
```

### The model accuracy on test data is 75% i.e the model prediction is correct 75 times out of 100

```python
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix(testoutput,y_pred_test), annot = True, fmt = 'd')
plt.xlabel("Actual")
plt.ylabel("Prediction")
```

# ARTIFICIAL NEURAL NETWORK

```python
import tensorflow as tf
from tensorflow import keras

model= keras.Sequential([
    keras.layers.Dense(20, input_shape=(37,), activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])
model.compile(optimizer ='adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
model.fit(X_train,y_train, epochs = 100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred_ANN = model.predict(X_test)
y_pred_ANN
```

```python
y_p_ANN= []
for i in y_pred_ANN:
    if i>0.5:
        y_p_ANN.append(1)
    else:
        y_p_ANN.append(0)
y_p_ANN[:5]
```

```python
accuracy_score(y_test,y_p_ANN)
```

```python
print(classification_report(y_test,y_p_ANN))
```

```python
y_pred_test_ANN = model.predict(testdata)
y_p_test_ANN = []
for i in y_pred_test_ANN:
    if i > 0.5:
        y_p_test_ANN.append(1)
    else:
        y_p_test_ANN.append(0)
    
accuracy_score(testoutput,y_p_test_ANN)
```

```python
print(classification_report(testoutput,y_p_test_ANN))
```

### The ANN model accuracy on test data is 85% i.e the model prediction is correct 85 times out of 100

```python
sns.heatmap(confusion_matrix(testoutput,y_p_test_ANN), fmt = 'd', annot = True, cmap= 'RdYlGn')
```
